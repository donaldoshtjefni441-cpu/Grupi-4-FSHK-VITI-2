```python
import numpy as np
import matplotlib.pyplot as plt
import os
#Punuan:Donaldo Shtjefni,Erik Cupi,Klaus Dragjoshi
# Krijon folderin 'figures' nëse nuk ekziston që të mos japë gabim
if not os.path.exists('figures'):
    os.makedirs('figures')

# --- Detyra 3: Ushtrim me NumPy ---
t = np.linspace(0, 10, 1000)
# Funksioni i ri: f(t) = cos(t) + 0.3*sin(5t)
f = np.cos(t) + 0.3 * np.sin(5 * t)

# Llogaritjet numerike
dt = t[1] - t[0]
df = np.gradient(f, dt)
integral = np.cumsum(f) * dt
norma = np.linalg.norm(f)
vlera_mesatare = np.mean(f)
maksimumi = np.max(f)
minimumi = np.min(f)

print(f"Norma: {norma:.4f}")
print(f"Mesatarja: {vlera_mesatare:.4f}")
print(f"Max: {maksimumi:.4f}, Min: {minimumi:.4f}")

# --- Detyra 1: Figura e stilizuar ---
plt.figure(figsize=(10, 6))
plt.plot(t, f, label="f(t) = cos(t) + 0.3sin(5t)", color="red", linewidth=2)
plt.fill_between(t, f, color="red", alpha=0.1)

plt.title("Identiteti Grafik i Grupit", fontsize=14)
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid(True)
plt.legend()

# Tani kjo nuk do dështojë
plt.savefig("figures/signature_plot.png")
plt.show()
```
